```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# Calculate the total number of students
total_students = school_data_complete["student_name"].count()

# Calculate the number of students who passed math (math scores greater than or equal to 70)
passing_math_count = school_data_complete[school_data_complete["math_score"] >= 70].count()["student_name"]

# Calculate the percentage of students who passed math
passing_math_percentage = passing_math_count / total_students * 100

# Print the result
print("Percentage of students passing math:", passing_math_percentage)
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
total_students = len(school_data_complete)
passing_reading_count = school_data_complete[school_data_complete["reading_score"] >= 70].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(total_students) * 100
print("Percentage of students passing reading:", passing_reading_percentage)
```

```python
# Use the following to calculate the percentage of students that passed math and reading
# Calculate the total number of students
total_students = school_data_complete["Student ID"].nunique()

# Calculate the number of students passing both math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["Student ID"]

# Calculate the overall passing rate
overall_passing_rate = passing_math_reading_count / total_students * 100

# Print the result
print("Overall Passing Rate:", overall_passing_rate)
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({
    "Total Schools": [school_data['school_name'].nunique()],
    "Total Students": [student_data['student_name'].count()],
    "Total Budget": [school_data['budget'].sum()],
    "Average Math Score": [student_data['math_score'].mean()],
    "Average Reading Score": [student_data['reading_score'].mean()],
    "% Passing Math": [(student_data['math_score'] >= 70).mean() * 100],
    "% Passing Reading": [(student_data['reading_score'] >= 70).mean() * 100],
    "% Overall Passing": [(student_data[(student_data['math_score'] >= 70) & (student_data['reading_score'] >= 70)].count()['student_name'] / student_data['student_name'].count()) * 100]
})

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```
